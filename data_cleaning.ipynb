```python
import pandas as pd

df = pd.read_excel("Online Retail.xlsx")

print(df.shape)
```

```python
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['Customer ID'])

print("After removing missing Customer ID:", df.shape)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print(df['InvoiceDate'].dtype)  # should show: datetime64
```

```python
df = df[df['Quantity'] > 0]

print("After removing bad Quantity rows:", df.shape)
```

```python
df = df[df['Price'] > 0]

print("After removing bad UnitPrice rows:", df.shape)
```

```python
df['Revenue'] = df['Quantity'] * df['Price']

print(df[['Quantity', 'Price', 'Revenue']].head())
```

```python
print("Final dataset shape:", df.shape)
print(df.head())
print(df.isnull().sum())  # should show 0 or very few nulls now
```

```python
# Rename and save as CSV
df.to_csv("Online_Retail_Cleaned.csv", index=False)
print("Cleaned file saved successfully!")
```
